```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Preprocesar los datos
data = df[['close']].values
scaler = MinMaxScaler(feature_range=(0, 1))
data = scaler.fit_transform(data)

# Dividir los datos en conjuntos de entrenamiento y prueba
train_size = int(len(data) * 0.9)
train_data, test_data = data[0:train_size,:], data[train_size:len(data),:]

# Crear el conjunto de datos para LSTM
def create_dataset(dataset, look_back=1):
    X, Y = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        X.append(a)
        Y.append(dataset[i + look_back, 0])
    return np.array(X), np.array(Y)

look_back = 10
X_train, Y_train = create_dataset(train_data, look_back)
X_test, Y_test = create_dataset(test_data, look_back)

# Remodelar la entrada para que sea [muestras, pasos de tiempo, características]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Crear y entrenar el modelo LSTM
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, Y_train, epochs=70, batch_size=32)

# Hacer predicciones
predictions = model.predict(X_test)

# Invertir la escala para obtener las predicciones en la escala original
predictions = scaler.inverse_transform(predictions)
Y_test = scaler.inverse_transform(Y_test.reshape(-1, 1))

print(predictions.shape)
print(Y_test.shape  )

# evaluar el modelo
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Calculate MSE,RMSE, MAE,R-squared
mse = mean_squared_error(Y_test, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(Y_test, predictions)
r2 = r2_score(Y_test, predictions)

# Print the results
print('MSE:', mse)
print('RMSE:', rmse)
print('MAE:', mae)
print('R-squared:', r2)
```

```python
def get_performance(last_2_days_comparison_df, threshold_percentage):

    initial_capital = 10000
    capital = initial_capital
    #threshold_percentage = percentage    #0.002 # 0.2%

    print(f"Starting simulation with initial capital: ${capital:.2f}")

    # Iterate through the last 10 days of data
    # We need the actual price of the current day and the predicted price for the next day
    # The DataFrame index is the date

    # We will start from the second day in the last_10_days_comparison_df to have a 'previous' day's actual price
    for i in range(1, len(last_2_days_comparison_df)):
        previous_day_actual_price = last_2_days_comparison_df['Actual Price'].iloc[i-1]
        current_day_predicted_price = last_2_days_comparison_df['Predicted Price'].iloc[i]
        current_day_actual_price = last_2_days_comparison_df['Actual Price'].iloc[i]
        current_date = last_2_days_comparison_df.index[i].date() # Get the date for printing

        # Calculate the price change percentage based on the prediction
        predicted_change_percentage = (current_day_predicted_price - previous_day_actual_price) / previous_day_actual_price

        # Implement the trading strategy
        action = "Hold" # Default action

        if predicted_change_percentage > threshold_percentage:
            action = "Buy"
            # Simulate buying: Assume we buy at the previous day's closing price and sell at the current day's closing price
            # This is a simplified simulation. A real strategy would involve more complex order execution.
            # Calculate the percentage change based on actual prices for the day
            actual_change_percentage = (current_day_actual_price - previous_day_actual_price) / previous_day_actual_price
            capital *= (1 + actual_change_percentage)
            print(f"{current_date}: Predicted increase ({predicted_change_percentage:.4f}). Action: {action}. Capital after trading: ${capital:.2f}")

        elif predicted_change_percentage < -threshold_percentage:
            action = "Sell"
            # Simulate selling/shorting: Assume we benefit from a price decrease
            # This is a simplified simulation of profiting from a price drop.
            # In a real shorting scenario, the profit is the difference between the short price and the buy-back price.
            # For simplicity here, we'll apply the inverse of the actual change percentage to the capital.
            actual_change_percentage = (current_day_actual_price - previous_day_actual_price) / previous_day_actual_price
            capital *= (1 - actual_change_percentage) # Simulating profit from price decrease
            print(f"{current_date}: Predicted decrease ({predicted_change_percentage:.4f}). Action: {action}. Capital after trading: ${capital:.2f}")
        else:
            # No action if the predicted change is within the threshold
            print(f"{current_date}: Predicted change ({predicted_change_percentage:.4f}) within threshold. Action: {action}. Capital remains: ${capital:.2f}")


    # Calculate the total percentage of benefits
    total_benefit_percentage = ((capital - initial_capital) / initial_capital) * 100
    return total_benefit_percentage
    print(f"\nFinal Capital after 1 day simulation: ${capital:.2f}")
    print(f"Total Percentage of Benefits: {total_benefit_percentage:.2f}%")
```
